```python
import random
import matplotlib.pyplot as plt


# -----------------------------
# Interpolation Search (Float)
# -----------------------------
def interpolation_search_float(arr, target):

    low = 0
    high = len(arr) - 1
    comparisons = 0

    while (
        low <= high
        and target >= arr[low]
        and target <= arr[high]
    ):

        comparisons += 1

        if arr[low] == arr[high]:
            if abs(arr[low] - target) < 1e-9:
                return low, comparisons
            return -1, comparisons

        pos = low + int(
            (
                (target - arr[low])
                * (high - low)
            )
            /
            (
                arr[high] - arr[low]
            )
        )

        if pos < low or pos > high:
            break

        if abs(arr[pos] - target) < 1e-6:
            return pos, comparisons

        elif arr[pos] < target:
            low = pos + 1

        else:
            high = pos - 1

    return -1, comparisons


# -----------------------------
# Dataset Experiment
# -----------------------------
sizes = [10000, 50000, 100000]

comparison_results = []

print("\nFLOAT INTERPOLATION SEARCH ANALYSIS\n")

for size in sizes:

    data = sorted(
        random.uniform(0.0, 1000.0)
        for _ in range(size)
    )

    target = random.choice(data)

    index, comparisons = (
        interpolation_search_float(
            data,
            target
        )
    )

    comparison_results.append(comparisons)

    print(
        f"Size={size} | "
        f"Comparisons={comparisons}"
    )


# -----------------------------
# Chart
# -----------------------------
plt.figure(figsize=(8, 5))

plt.plot(
    sizes,
    comparison_results,
    marker="o",
    linewidth=3
)

plt.title(
    "Interpolation Search on Floating-Point Dataset"
)

plt.xlabel(
    "Dataset Size"
)

plt.ylabel(
    "Number of Comparisons"
)

plt.grid(True)

plt.show()
```
