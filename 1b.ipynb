```python
def interpolation_search(arr, target):
    low = 0
    high = len(arr) - 1
    probes = 0

    while low <= high and target >= arr[low] and target <= arr[high]:
        probes += 1

        # Avoid division by zero
        if arr[low] == arr[high]:
            if arr[low] == target:
                return low, probes
            return -1, probes

        # Interpolation formula
        pos = low + ((target - arr[low]) * (high - low)) // (arr[high] - arr[low])

        if arr[pos] == target:
            return pos, probes

        elif arr[pos] < target:
            low = pos + 1

        else:
            high = pos - 1

    return -1, probes


def binary_search(arr, target):
    low = 0
    high = len(arr) - 1
    probes = 0

    while low <= high:
        probes += 1
        mid = (low + high) // 2

        if arr[mid] == target:
            return mid, probes

        elif arr[mid] < target:
            low = mid + 1

        else:
            high = mid - 1

    return -1, probes


# Input roll numbers
n = int(input("Enter number of students: "))

print("Enter sorted roll numbers:")
roll_numbers = []

for i in range(n):
    roll_numbers.append(int(input()))

target = int(input("Enter roll number to search: "))


# Interpolation Search
i_index, i_probes = interpolation_search(roll_numbers, target)

# Binary Search
b_index, b_probes = binary_search(roll_numbers, target)


# Results
print("\n--- Results ---")

if i_index != -1:
    print("Interpolation Search: Found at index", i_index)
else:
    print("Interpolation Search: Not Found")

print("Interpolation Search Probes:", i_probes)


if b_index != -1:
    print("Binary Search: Found at index", b_index)
else:
    print("Binary Search: Not Found")

print("Binary Search Probes:", b_probes)


# Comparison
print("\nComparison:")

if i_probes < b_probes:
    print("Interpolation Search used fewer probes.")
elif b_probes < i_probes:
    print("Binary Search used fewer probes.")
else:
    print("Both used same number of probes.")
```
